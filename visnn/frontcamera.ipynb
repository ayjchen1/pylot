```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

%pwd
```

# Original Data (last week)

```python
df = pd.read_csv("C:\\Users\\jerry\\pylot\\visnn\\vis_data\\vis00.csv")
inputs = df.iloc[:, -4:-1]
outputs = df.iloc[:, -1]
```

```python
df
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#n = len(inputs.index)
n = 5000

for i in range(n):
    x = inputs.iloc[i, :].X
    y = inputs.iloc[i, :].Y
    error = outputs[i]
    ax.scatter(x, y, error, marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

## Original Data with outliers removed

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#n = len(inputs.index)
n = 5000

cnt = 0
for i in range(n):
    x = inputs.iloc[i, :].X
    y = inputs.iloc[i, :].Y
    error = outputs[i]
    if (error > 1000):
        continue
    if (error > 300):
        cnt += 1
    ax.scatter(x, y, error, marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

print(cnt)

plt.show()
```

## Original Data with only FRONT camera

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#n = len(inputs.index)
n = 4000

cnt = 0
for i in range(n):
    x = inputs.iloc[i, :].X
    y = inputs.iloc[i, :].Y
    error = outputs[i]
    if (error > 1000):
        continue
    if (error > 300):
        cnt += 1
    ax.scatter(x, y, error, marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

print(cnt)

plt.show()
```

```python
df = df.reset_index(drop=True)
inputs = df
outputs = df.iloc[:, -1]
inputs
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

#n = len(inputs.index)
n = 2000

for i in range(n):
    x = inputs.iloc[i, :].X
    y = inputs.iloc[i, :].Y
    error = outputs[i]
    if (df.iloc[i, :].CAMERA=="FRONT-eval"):
        color = "#e06e5c" if (error > 1000) else "#88c41f"
    else: # back camera
        color = "#a8210c" if (error > 1000) else "#537a0d" # darker version of red/green
        
    ax.scatter(x, y, c=color, marker='o')

ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

# Front Camera

```python
fig = plt.figure()
ax = fig.add_subplot(111)

#n = len(inputs.index)
n = 5000

for i in range(n):
    x = inputs.iloc[i, :].X
    y = inputs.iloc[i, :].Y
    error = outputs[i]
    if (error > 1000):
        ax.scatter(x, y, c="#e06e5c", marker='o')
    else:
        ax.scatter(x, y, c="#88c41f", marker='v')

ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```
