# Original Data (last week)

```python
df = pd.read_csv("C:\\Users\\jerry\\pylot\\visnn\\vis_data\\newvis.csv")
inputs = df.iloc[:, -4:-1]
outputs = df.iloc[:, -1]
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#n = len(inputs.index)
n = 5000

for i in range(n):
    x = inputs.iloc[i, :].X
    y = inputs.iloc[i, :].Y
    error = outputs[i]
    ax.scatter(x, y, error, marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

plt.show()
```

## Original Data with outliers removed

```python
len(inputs.index)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

#n = len(inputs.index)
n = len(inputs.index)

cnt = 0
for i in range(n):
    x = inputs.iloc[i, :].X
    y = inputs.iloc[i, :].Y
    error = outputs[i]
    if (error > 1000):
        continue
    if (error > 300):
        cnt += 1
    ax.scatter(x, y, error, marker='o')

ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

print(cnt)
#ax.set_xlim([-80, 80])
#ax.set_ylim([-80, 80])

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111)

#n = len(inputs.index)
n = 5000

for i in range(n):
    x = inputs.iloc[i, :].X
    y = inputs.iloc[i, :].Y
    error = outputs[i]
    if (error > 1000):
        ax.scatter(x, y, c="#e06e5c", marker='o')
    else:
        ax.scatter(x, y, c="#88c41f", marker='v')

ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```
